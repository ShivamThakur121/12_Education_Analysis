## Application and Enrollment Analysis

### Average number of applications received by colleges

```python
average_applications = data['Apps'].mean()
average_applications
```

### Average acceptance rate

```python
average_acceptance_rate = (data['Accept'].sum() / data['Apps'].sum()) * 100
average_acceptance_rate
```

### Average enrollment rate

```python
average_enrollment_rate = (data['Enroll'].sum() / data['Accept'].sum()) * 100
average_enrollment_rate
```

### College with the highest number of applications received

```python
college_highest_applications = data.loc[data['Apps'].idxmax()]['Names']
college_highest_applications
```
